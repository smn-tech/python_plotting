```python
import numpy as np
import matplotlib.pyplot as plt
```

**Create data**

```python
x=np.linspace(-np.pi,np.pi,100)

y1=np.sin(x)
y2=np.cos(x)
```

**Plotting section**

```python
plt.figure(figsize=(12,3))

plt.plot(x,y1,label='Sin')
plt.plot(x,y2,label='Cos')
plt.legend()
plt.grid()

plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Title of the figure')


plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
plt.plot(x,y1,label='Sin')
plt.plot(x,y2,label='Cos')
plt.legend()
plt.grid()

plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Title of the figure-1')


plt.subplot(2,1,2)
plt.plot(x,y1,label='Sin')
plt.plot(x,y2,label='Cos')
plt.legend()
plt.grid()

plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Title of the figure-2')

plt.tight_layout()

plt.show()
```
